```python
# The Usual
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
listings = pd.read_csv("listings.csv", usecols=['id', 'name', 'description', 'neighbourhood_cleansed', 'zipcode', 'property_type', 'room_type', 
                                            'price', 'availability_365'])
listings = listings.rename(columns={'id': 'listing_id'})
```

```python
neighbourhood = pd.DataFrame()
neighbourhood['neighbourhood'] = listings['neighbourhood_cleansed'].unique()

# neighbourhood['Count'] = ""
neighbourhood['Count'] = listings.groupby(['neighbourhood_cleansed'])['listing_id'].transform('count')
```

```python
neighbourhood2 = pd.DataFrame()
neighbourhood2['neighbourhood'] = listings['neighbourhood_cleansed'].unique()

# neighbourhood['Count'] = ""
neighbourhood2['Count'] = listings.groupby(['neighbourhood_cleansed'])['listing_id'].transform('count')
```

```python
neighbourhood
```

```python
# Get names of indexes for which column Age has value 30
indexNames = neighbourhood[neighbourhood['Count'] <= 3800].index
 
# Delete these row indexes from dataFrame
neighbourhood.drop(indexNames , inplace=True)
```

```python
plt.figure(figsize=(26,12))
plt.xlabel('Neighbourhood')
plt.ylabel('Frequency')
plt.title('Most Common Neighbourhoods')
plt.bar(neighbourhood['neighbourhood'], neighbourhood['Count'], color='royalblue')
plt.xticks(rotation=45)

plt.show()
```

```python
# Get names of indexes for which column Age has value 30
indexNames2 = neighbourhood2[neighbourhood2['Count'] >= 100].index
 
# Delete these row indexes from dataFrame
neighbourhood2.drop(indexNames2, inplace=True)
```

```python
plt.figure(figsize=(26,12))
plt.xlabel('Neighbourhood')
plt.ylabel('Frequency')
plt.title('Least Common Neighbourhoods')
plt.bar(neighbourhood2['neighbourhood'], neighbourhood2['Count'], color='royalblue')
plt.xticks(rotation=45)

plt.show()
```
